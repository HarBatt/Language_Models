```python
import pandas as pd
```

```python
n = 60
dataframe = pd.read_csv("data.csv")
dataframe.head()
```

```python
category = dataframe['category'].value_counts()[:n].index.tolist()
```

```python
dataframe_new = []
for index, row in dataframe.iterrows():
    if row["category"] in category:
        dataframe_new.append(row)
        
df = pd.DataFrame(dataframe_new)
```
